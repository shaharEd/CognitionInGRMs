## Demonstration of structure
The model is of the form $\frac{dM}{dt} = F(M)-\mu_M M + I_ext$, 
which creates positive feedback (F is a nonlinear, similar to sigmoid, monotonic in M) with saturation - F goes to saturation which makes sure nothing is exploding (similar to the $\frac{dx}{dt} = \beta - \delta x$ we saw in class) 
This creates three fixed points, one in 0, one in 1 (both stable) and on unstable in the middle.

We use stochastic dynamics, so when I is in the middle, the neuron is likely to maintain ("remember") its state. When I is high, the neuron is likely to change to 1, when I is low, to 0. 


The model is like this: with no US, I is in the mid-range (leading the neuron to likely stay in its state). When US appears, if CS appears I is high ("Learning" - neuron -> 1), if CS no appearing, I gets low, leading the neuron to 0 ("Extintion").

# Retrival
$\frac{dR}{dt} = -\mu_R R + G_1 (U) + G_2 (C,M)$ Where $ -\mu_R R$ controls the explotion, and $ G_1 , G_2$ are non-linear sigmoid-like functions. $G_1$ is monotonic with $U$, and $G_2$ is high when $C \land M$ exists

> We should note that the period of the extinction and learning in the model is stochastic. That is, every appearance of the US without the CS is like an experiment in which the dynamics of M might go over the bifurcation point in the same chance. 

# Multi-CS

The idea is the a CS would be a combination of $C_i$ s. These are the A B C D E in the original paper. Then, the authors present the order effect, or the effect of forgetting; a pairing between a CS and US might be forgotten if other pairing have already happened (beacause the $M_i$s will likely to be updated to the match the new CS that was presented). Denote $p, q$ the probabilities that after seeing a pair of CS/US the $M_i$s change from low-to-high (when $C_i$ is activated in the CS, $p$) and from high-to-low (when $C_i$ is deactivated in the CS, $q$). If $p=q=1$, the capacity is 1 - after seeing a new pairing, all other CSs are indistinguishable, no matter if they were watched the step before. If $p,q$ are low, we'll not really learn and would likely have capacity of 0. Therefore, we have to optimize $p,q$ to reach highest capacity. 
Then, we can measure the signal-to-noise ratio between the $v^{th}$ memory (in a reveresed order, the most recent is v=1) and its SNR, that is, the distribution of the response when it is appears alone (that is, only the CS), and the distribution of the not-learned stimulus. We calculate it by $\frac{\mu_{v^{th}} - \mu_{null}}{\sqrt{\sigma_{v^{th}} ^ 2 + \sigma_{null} ^ 2}}$
